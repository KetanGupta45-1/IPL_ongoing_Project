```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv("matches_added_averages.csv")

# --- Add Match_ID ---
match_id = 0
match_ids = []
for _, row in df.iterrows():
    if row['Over'] == 0.0:
        match_id += 1
    match_ids.append(match_id)

df['Match_ID'] = match_ids

# Step 1: Categorize over phase
def categorize_over(over):
    over_num = int(str(over).split('.')[0])
    if over_num < 6:
        return 1  # Powerplay
    elif over_num < 16:
        return 2  # Middle
    else:
        return 3  # Death

df['Over_Phase'] = df['Over'].apply(categorize_over)

# Step 2: Cumulative team runs before the ball
df['Cumulative_Runs'] = df.groupby(['Match_ID', 'Team_ID'])['Total_Runs'].cumsum().shift(fill_value=0)

# Step 3 & 4: Wicket loss + cumulative wickets + partnership runs
df['Is_Wicket'] = (df['Total_Runs'] == 0).astype(int)  # Proxy for wicket
df['Cumulative_Wickets'] = df.groupby(['Match_ID', 'Team_ID'])['Is_Wicket'].cumsum()

df['Partnership_Runs'] = 0
last_wicket_indices = {}

for idx, row in df.iterrows():
    key = (row['Match_ID'], row['Team_ID'])
    if key not in last_wicket_indices:
        last_wicket_indices[key] = -1
    if row['Is_Wicket'] == 1:
        last_wicket_indices[key] = idx
    start = last_wicket_indices[key] + 1
    df.at[idx, 'Partnership_Runs'] = df.loc[start:idx, 'Total_Runs'].sum()

df['Partnership_Runs'] = df['Partnership_Runs'].astype(int)

# Step 5: Pressure Index (PI)
INITIAL_RR = 10
WICKET_WEIGHT = 25
TOTAL_BALLS = 120
BENCHMARK_TARGET = 190  # midpoint of 180-200

# Detect innings using match-wise logic
df['Innings'] = df.groupby('Match_ID')['Team_ID'].transform(lambda x: (x != x.shift()).cumsum())

# Compute target scores
target_map = df[df['Innings'] == 1].groupby('Match_ID')['Total_Runs'].sum() + 1
df['Target'] = df['Match_ID'].map(target_map).where(df['Innings'] == 2, BENCHMARK_TARGET)

# Ball tracking
df['Ball_Count'] = df.groupby(['Match_ID', 'Team_ID']).cumcount() + 1
df['Balls_Left'] = TOTAL_BALLS - df['Ball_Count']

# Current and required RR
df['Current_RR'] = df['Cumulative_Runs'] / (df['Ball_Count'] / 6)
df['Req_RR'] = np.where(
    df['Innings'] == 2,
    (df['Target'] - df['Cumulative_Runs']) / (df['Balls_Left'] / 6),
    BENCHMARK_TARGET / 20
)

# Pressure Index formula
df['PI'] = (
    (df['Req_RR'] / INITIAL_RR) * 100 +
    ((WICKET_WEIGHT * df['Cumulative_Wickets']) / (180 * df['Target'])) *
    (df['Balls_Left'] / TOTAL_BALLS) *
    ((df['Target'] - df['Cumulative_Runs']) / df['Target'])
)

# Step 6: Dynamic Pitch Change - rolling RR over 12 balls
df['Rolling_Runs'] = df.groupby(['Match_ID', 'Team_ID'])['Total_Runs'].rolling(window=12, min_periods=1).mean().reset_index(level=[0,1], drop=True) * 6
df['Previous_Rolling_Runs'] = df['Rolling_Runs'].shift().fillna(method='bfill')
df['Pitch_Change'] = df['Rolling_Runs'] - df['Previous_Rolling_Runs']

# Optional: Save enhanced data
df.to_csv("matches_enhanced_with_features.csv", index=False)
```
