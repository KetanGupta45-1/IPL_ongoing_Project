```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv('match_data_23-24.csv')
df.head(10)
```

```python
df['Venue'].unique()
```

```python
venue_name = ['Narendra Modi Stadium, Ahmedabad',
       'Punjab Cricket Association IS Bindra Stadium, Mohali, Chandigarh',
       'Bharat Ratna Shri Atal Bihari Vajpayee Ekana Cricket Stadium, Lucknow',
       'Rajiv Gandhi International Stadium, Uppal, Hyderabad',
       'M Chinnaswamy Stadium, Bengaluru',
       'MA Chidambaram Stadium, Chepauk, Chennai',
       'Arun Jaitley Stadium, Delhi',
       'Barsapara Cricket Stadium, Guwahati', 'Eden Gardens, Kolkata',
       'Wankhede Stadium, Mumbai', 'Sawai Mansingh Stadium, Jaipur',
       'Himachal Pradesh Cricket Association Stadium, Dharamsala',
       'Maharaja Yadavindra Singh International Cricket Stadium, Mullanpur',
       'Dr. Y.S. Rajasekhara Reddy ACA-VDCA Cricket Stadium, Visakhapatnam']
```

```python
output_dir = "per_venue_1st_innings_summary"
os.makedirs(output_dir, exist_ok=True)
```

```python
for venue in venue_name:
    df_venue = df[df["Venue"] == venue]

    # List to store summary rows
    summary_rows = []

    # Loop through matches at this venue
    for match_date in df_venue["Date"].unique():
        df_match = df_venue[df_venue["Date"] == match_date]

        # Find first row where Innings No. == 2
        second_innings_start = df_match[df_match["Innings No."] == 2].head(1)
        if not second_innings_start.empty:
            first_innings_team = df_match[df_match["Innings No."] == 1]["Batting Team"].iloc[0]
            first_innings_total = df_match[df_match["Innings No."] == 1]["Cumulative Team Runs"].max()
            bowling_team = df_match[df_match["Innings No."] == 1]["Bowling Team"].iloc[0]

            summary_rows.append({
                "Date": match_date,
                "Venue": venue,
                "Batting Team": first_innings_team,
                "Bowling Team": bowling_team,
                "1st Innings Total": first_innings_total
            })

    # If there's any data, save to CSV
    if summary_rows:
        df_summary = pd.DataFrame(summary_rows)
        filename = venue.replace(",", "").replace(" ", "_") + ".csv"
        df_summary.to_csv(os.path.join(output_dir, filename), index=False)
        
```

```python
folder_path = "per_venue_1st_innings_summary"
dict_of_scores = {}

for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(folder_path, filename)
        df_venue = pd.read_csv(file_path)

        target_list = []
        past_scores = []

        for score in df_venue["1st Innings Total"]:
            if len(past_scores) == 0:
                target_list.append(np.nan)
            elif len(past_scores) == 1:
                target_list.append(past_scores[0])
            else:
                target_list.append(np.mean(past_scores))

            if not np.isnan(score):
                past_scores.append(score)

        venue_name = filename.replace(".csv", "").replace("_", " ")
        dict_of_scores[venue_name] = target_list

print(dict_of_scores)
```
