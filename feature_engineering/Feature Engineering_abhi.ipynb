### Match ID

```python
# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract year
df['Year'] = df['Date'].dt.year

# Create a unique integer Match ID using factorize on (Year, Match No.)
df['Match ID'], _ = pd.factorize(df[['Year', 'Match No.']].apply(tuple, axis=1))

df
```

```python
df.drop(columns=['Year'], inplace=True)
df.columns
```

### is_wicket

```python
df['pair'] = df[['Striker', 'Non Striker']].apply(lambda row: tuple(sorted(row)), axis=1)

df['prev_pair'] = df['pair'].shift(1)
df['prev_innings'] = df['Innings No.'].shift(1)

df['is_wicket'] = (
    (df['Innings No.'] == df['prev_innings']) &
    (df['pair'] != df['prev_pair'])
).astype(int)
```

```python
df
```

```python
df.drop(columns=['pair', 'prev_pair', 'prev_innings'], inplace=True)
```

### Total Wickets

```python
df['Total Wickets'] = df.groupby(['Match ID', 'Innings No.'])['is_wicket'].cumsum()
```

```python
df
```

### Balls Count & Balls Left

```python
def is_legal(extras_dict):
    return not any(k in extras_dict for k in ['wides', 'noballs'])

df['is_legal_delivery'] = df['Extras'].apply(is_legal)

df['Balls Count'] = df.groupby(['Match ID', 'Innings No.'])['is_legal_delivery'].cumsum()
df.head(130)
```

```python
df.drop(columns=['is_legal_delivery'], inplace=True)
```

```python
df["Balls Left"] = 120 - df["Balls Count"]
```

```python
df
```

### Runs on Previous ball

```python
# Group by Match ID and Innings No., and shift the desired columns
df['Batsman Runs on Previous Ball'] = df.groupby(['Match ID', 'Innings No.'])['Runs by Batsman'].shift(1)
df['Total Runs on Previous Ball'] = df.groupby(['Match ID', 'Innings No.'])['Total Runs (Ball)'].shift(1)

# Optional: fill NaNs (first ball of innings) with 0
df['Batsman Runs on Previous Ball'].fillna(0, inplace=True)
df['Total Runs on Previous Ball'].fillna(0, inplace=True)

df
```

```python
df['Batsman Runs on Previous Ball'] = df['Batsman Runs on Previous Ball'].fillna(0).astype(int)
df['Total Runs on Previous Ball'] = df['Total Runs on Previous Ball'].fillna(0).astype(int)
```

```python
df
```

### Over Phase

```python
def categorize_over(over):
    over_num = int(str(over).split('.')[0])
    if over_num < 6:
        return 1  # Powerplay
    elif over_num < 16:
        return 2  # Middle
    else:
        return 3  # Death

df['Over_Phase'] = df['Over'].apply(categorize_over)
```

```python
df
```

### Player Strike Rate

```python
# (Optional) Define legal delivery if not already done
def is_legal(extras_dict):
    return not any(k in extras_dict for k in ['wides', 'noballs'])

df['is_legal_delivery'] = df['Extras'].apply(is_legal)

df['Balls Faced'] = df.groupby(['Match ID', 'Innings No.', 'Striker'])['is_legal_delivery'].cumsum()

# Calculate Strike Rate (and round)
df['Striker Strike Rate'] = (df['Cumulative Runs by Batsman'] / df['Balls Faced']) * 100

# Fill NA values in Strike Rate with the previous entry
df['Striker Strike Rate'] = df['Striker Strike Rate'].fillna(method='ffill').round(2)

df
```

```python
df.drop(columns = ['Balls Faced'], inplace=True)
```

### Team Strike Rate

```python
df['Team Strike Rate'] = (df['Cumulative Team Runs'] / df['Balls Count']) * 100

# Fill NA values in Strike Rate with the previous entry
df['Team Strike Rate'] = df['Team Strike Rate'].fillna(method='ffill').round(2)
```

```python
df
```

### Partnership Runs

```python
# Step 1: Define a unique partnership ID (reset on wicket)
df['Partnership ID'] = df.groupby(['Match ID', 'Innings No.'])['is_wicket'].cumsum()

# Step 2: Compute running sum of runs in the partnership
df['Partnership Runs'] = df.groupby(['Match ID', 'Innings No.', 'Partnership ID'])['Runs by Batsman'].cumsum()
df
```

```python
df.drop(columns = ['Partnership ID'], inplace=True)
```

### Current Runrate

```python
df["Current Run Rate"] = (df['Cumulative Team Runs']/df['Balls Count']) * 100
```

```python
df["Current Run Rate"] = df['Current Run Rate'].round(2)
```

```python
df
```

## Dot Ball

```python
for i in range(len(df)):
    if df.loc[i, 'is_legal_delivery'] and df.loc[i, 'Runs by Batsman'] == 0:
        df.loc[i, 'is_dot'] = 1
    else:
        df.loc[i, 'is_dot'] = 0

df.is_dot = df.is_dot.fillna(0).astype(int)
```

```python
df['Dot Count'] = df.groupby(['Match ID', 'Innings No.'])['is_dot'].cumsum()
df['Dot Count'] = df['Dot Count'].fillna(0).astype(int)
df
```

```python
df.drop(columns = ['is_dot'], inplace=True)
```

### Boundary Count

```python
for i in range(len(df)):
    if df.loc[i, 'is_legal_delivery'] and (df.loc[i, 'Runs by Batsman'] == 4 or df.loc[i, 'Runs by Batsman'] == 6):
        df.loc[i, 'is_boundary'] = 1
    else:
        df.loc[i, 'is_boundary'] = 0

df.is_boundary = df.is_boundary.fillna(0).astype(int)
```

```python
df['Boundary Count'] = df.groupby(['Match ID', 'Innings No.'])['is_boundary'].cumsum()
df['Boundary Count'] = df['Boundary Count'].fillna(0).astype(int)
df
```

### Average Runs by Striker

```python
runs_buffer = []

for i in range(len(df)):
    if df.loc[i, 'Over'] == 0.0:
        runs_buffer = []
        df.loc[i, 'Average Striker Runs 3'] = 0.0
    else:
        if len(runs_buffer) > 0:
            avg = sum(runs_buffer) / len(runs_buffer)
        else:
            avg = 0.0
        df.loc[i, 'Average Striker Runs 3'] = avg

    runs_buffer.append(df.loc[i, 'Runs by Batsman'])

    if len(runs_buffer) > 3:
        runs_buffer.pop(0)
```

### Average Runs Conceded by Bowler

```python
runs_buffer = []

for i in range(len(df)):
    if df.loc[i, 'Over'] == 0.0:
        runs_buffer = []
        df.loc[i, 'Average Conceded Runs by Bowler 6'] = 0.0
    else:
        if len(runs_buffer) > 0:
            avg = sum(runs_buffer) / len(runs_buffer)
        else:
            avg = 0.0
        df.loc[i, 'Average Conceded Runs by Bowler 6'] = avg

    runs_buffer.append(df.loc[i, 'Runs Conceded by Bowler'])

    if len(runs_buffer) > 6:
        runs_buffer.pop(0)
```

### Target

```python
# Get the target for each match (max Cumulative Team Runs from 1st innings)
first_innings = df[df["Innings No."] == 1]
targets = first_innings.groupby("Match ID")["Cumulative Team Runs"].max().rename("Target")

# Merge the targets back into the main dataframe
df = df.merge(targets, on="Match ID", how="left")

# Assign target only to 2nd innings; set as NaN for 1st innings
df["Target"] = df.apply(lambda row: row["Target"] if row["Innings No."] == 2 else None, axis=1)

df
```

```python
# Step 1: Create actual targets for 2nd innings
first_innings = df[df["Innings No."] == 1]
actual_targets = first_innings.groupby("Match ID")["Cumulative Team Runs"].max().rename("Actual_Target")
df = df.merge(actual_targets, on="Match ID", how="left")

# Step 2: Assign actual target to 2nd innings
df["Target"] = df.apply(
    lambda row: row["Actual_Target"] if row["Innings No."] == 2 else None,
    axis=1
)

# Step 3: Progressive average target for 1st innings (match-based)
venue_history = {}  # key: venue, value: list of past final 1st innings scores
match_targets = {}

# Get unique 1st innings Match IDs in order of appearance
match_ids = df[df["Innings No."] == 1]["Match ID"].drop_duplicates()

for match_id in match_ids:
    match_df = df[(df["Match ID"] == match_id) & (df["Innings No."] == 1)]
    venue = match_df["Venue"].iloc[0]
    final_score = match_df["Cumulative Team Runs"].max()

    # Get average of past scores at this venue
    past_scores = venue_history.get(venue, [])
    if past_scores:
        target = sum(past_scores) / len(past_scores)
    else:
        target = None  # First match at venue => NaN

    # Save target for this match
    match_targets[match_id] = target

    # Update venue history AFTER assigning target
    venue_history.setdefault(venue, []).append(final_score)

# Step 4: Assign 1st innings targets to all rows
for match_id, target in match_targets.items():
    df.loc[(df["Match ID"] == match_id) & (df["Innings No."] == 1), "Target"] = target

# Step 5: Drop helper column
df.drop(columns=["Actual_Target"], inplace=True)

# Optional: Save
# df.to_csv("match_with_targets.csv", index=False)
df[df['Venue'] == 'Narendra Modi Stadium, Ahmedabad']
```

### Pressure Index

```python

gamma = 0.2
# Compute the Pressure Index (PI) with dynamic α
df["Pressure Index"] = df.apply(
    lambda row: (
        # Adjust α based on Req. Run Rate
        (0.3 if row["Req. Run Rate"] <= 12 else
         (0.5 if row["Req. Run Rate"] <= 15 else
          (0.6 if row["Req. Run Rate"] <= 18 else 0.7))) *
        ((row["Req. Run Rate"] / (row["Current Run Rate"] if row["Current Run Rate"] != 0 else 0.0001)) *
         (row["Balls Count"] / 120))

        +

        # Adjust β based on Total Wickets
        (0.3 if row["Total Wickets"] < 4 else
         (0.4 if row["Total Wickets"] == 4 else
          (0.5 if row["Total Wickets"] == 5 else
           (0.7 if row["Total Wickets"] == 6 else 0.84)))) *
        (row["Total Wickets"] / 10)

        +
        (0.2 if row['Dot Count'] < 20 else
        (0.3 if row['Dot Count'] < 25 else
        (0.45 if row['Dot Count'] < 30 else
        (0.6 if row['Dot Count'] < 35 else
            (0.75 if row['Dot Count'] < 40 else 0.9))))) * row['Dot Count']

    ), axis=1
)
```

```python
mask = ((df["Innings No."] == 1) & (df["Over"] == 0.1)) | ((df["Innings No."] == 2) & (df["Over"] == 0.1))
df.loc[mask, "Pressure Index"] = df.loc[mask, "Pressure Index"].apply(lambda x: np.log2(x + 1))
```
